```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time

start_year = 2010  # Example start year
end_year = 2022    # Example end year
years = list(range(start_year, end_year + 1, 4))  # Every 4 years
all_matches = []

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

for year in years:
    url = f"https://fbref.com/en/comps/1/{year}-World-Cup-Stats"
    print(f"Fetching data for year: {year}")

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()
    except requests.RequestException as e:
        print(f"Request failed for {year}: {e}")
        continue

    soup = BeautifulSoup(response.text, 'html.parser')
    all_links = []

    # Find the specific table with the given ID using select
    tables = soup.select('table#results202210_overall')

    if not tables:
        print(f"No tables found for year: {year}")
        continue

    # Loop through each table (in case there are multiple matches)
    for table in tables:
        # Find all 'a' tags within this table
        links = table.find_all('a')
        # Extract the href attributes and filter them
        links = [l.get("href") for l in links if l.get("href") and '/squads/' in l.get("href")]
        # Construct the full URLs
        team_urls = [f"https://fbref.com{l}" for l in links]
        # Add to the list of all links
        all_links.extend(team_urls)

    if not all_links:
        print(f"No team links found for year: {year}")
        continue

    for team_url in team_urls:
        team_name = team_url.split("/")[-1].replace("-Stats", "").replace("-", " ")
        print(f"Processing team: {team_name} for year: {year}")

        for attempt in range(3):  # Retry up to 3 times
            try:
                data = requests.get(team_url, headers=headers)
                data.raise_for_status()
                break  # If request is successful, break out of the retry loop
            except requests.RequestException as e:
                if attempt == 2:  # If this was the last attempt, log the error
                    print(f"Request failed for team {team_name} after 3 attempts: {e}")
                else:
                    print(f"Request failed for team {team_name}, retrying... ({attempt+1}/3)")
                    time.sleep(10)  # Wait before retrying

        try:
            matches = pd.read_html(data.text, match="Scores & Fixtures")[0]
        except ValueError:
            print(f"No matches found for team {team_name}")
            continue

        soup = BeautifulSoup(data.text, 'html.parser')
        links = [l.get("href") for l in soup.find_all('a')]
        links = [l for l in links if l and 'all_comps/shooting/' in l]

        if not links:
            print(f"No shooting stats link found for team {team_name}")
            continue

        for attempt in range(3):  # Retry up to 3 times
            try:
                data = requests.get(f"https://fbref.com{links[0]}", headers=headers)
                data.raise_for_status()
                break  # If request is successful, break out of the retry loop
            except requests.RequestException as e:
                if attempt == 2:  # If this was the last attempt, log the error
                    print(f"Request failed for shooting stats of team {team_name} after 3 attempts: {e}")
                else:
                    print(f"Request failed for shooting stats of team {team_name}, retrying... ({attempt+1}/3)")
                    time.sleep(10)  # Wait before retrying

        try:
            shooting = pd.read_html(data.text, match="Shooting")[0]
        except ValueError:
            print(f"No shooting stats table found for team {team_name}")
            continue

        shooting.columns = shooting.columns.droplevel()

        try:
            team_data = matches.merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")
        except ValueError:
            print(f"Data merge failed for team {team_name}")
            continue

        team_data = team_data[team_data["Comp"] == "World Cup"]

        team_data["Season"] = year
        team_data["Team"] = team_name
        all_matches.append(team_data)
        time.sleep(5)  # Increase delay between team requests

if all_matches:
    match_df = pd.concat(all_matches)
else:
    print("No data collected.")

```
